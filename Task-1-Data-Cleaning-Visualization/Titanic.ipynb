1. Survival Distribution

Shows how many passengers survived.

```python
import matplotlib.pyplot as plt

survival_counts = titanic['Survived'].value_counts()

plt.figure(figsize=(6,4))
plt.bar(['Not Survived','Survived'], survival_counts)
plt.title('Survival Distribution')
plt.xlabel('Status')
plt.ylabel('Number of Passengers')
plt.show()
```

2. Gender Distribution

```python
gender_counts = titanic['Sex'].value_counts()

plt.figure(figsize=(6,4))
plt.bar(gender_counts.index, gender_counts.values)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

3. Age Distribution

```python
plt.figure(figsize=(8,5))
plt.hist(titanic['Age'].dropna(), bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

4. Passenger Class Distribution

```python
class_counts = titanic['Pclass'].value_counts().sort_index()

plt.figure(figsize=(6,4))
plt.bar(class_counts.index, class_counts.values)
plt.title('Passenger Class Distribution')
plt.xlabel('Class')
plt.ylabel('Number of Passengers')
plt.show()
```

5. Fare Distribution

```python
plt.figure(figsize=(8,5))
plt.hist(titanic['Fare'], bins=30)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

6. Survival by Gender

```python
survival_gender = pd.crosstab(titanic['Sex'], titanic['Survived'])

survival_gender.plot(kind='bar')
plt.title('Survival by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

8. Box Plot for Age (Outlier Detection)

```python
plt.figure(figsize=(6,4))
plt.boxplot(titanic['Age'].dropna())
plt.title('Boxplot of Age')
plt.ylabel('Age')
plt.show()
```

**INSIGHTS**

1.More passengers died than survived.

2.Male passengers were more than female passengers.

3.Most passengers were between 20 and 40 years old.

4.Most passengers belonged to 3rd class.

5.Most passengers paid lower fares.

6.Females had a higher survival rate than males.

7.Identifies age outliers and spread of ages.
