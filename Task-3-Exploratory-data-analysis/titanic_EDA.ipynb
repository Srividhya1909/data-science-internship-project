**Survival Percentage**

```python
import matplotlib.pyplot as plt
```

```python
survival_rate = titanic['Survived'].value_counts(normalize=True) * 100
print(survival_rate)
plt.figure(figsize=(6,6))
plt.pie(titanic['Survived'].value_counts(),
        labels=['Not Survived','Survived'],
        autopct='%1.1f%%')
plt.title('Survival Percentage')
plt.show()
```

**Passenger Class Distribution**

```python
class_counts = titanic['Pclass'].value_counts()

plt.bar(class_counts.index, class_counts.values)
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.title("Passenger Class Distribution")
plt.show()
```

**Gender Distribution**

```python
gender_counts = titanic['Sex'].value_counts()

plt.bar(gender_counts.index, gender_counts.values)
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

**Age Group Analysis**

```python
bins = [0,12,18,35,60,100]
labels = ['Child','Teen','Young Adult','Adult','Senior']

titanic['Age_Group'] = pd.cut(titanic['Age'], bins=bins, labels=labels)

print(titanic['Age_Group'].value_counts())

titanic['Age_Group'].value_counts().plot(kind='bar')
plt.title('Age Group Distribution')
plt.show()
```

**Fare Distribution**

```python
plt.figure(figsize=(8,5))
plt.hist(titanic['Fare'], bins=30)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

**Family Size Analysis**

```python
titanic['Family_Size'] = titanic['SibSp'] + titanic ['Parch'] + 1
plt.hist(titanic['Family_Size'], bins=10)
plt.title('Family Size Distribution')
plt.xlabel('Family Size')
plt.ylabel('Count')
plt.show()
```

**Survival by Embarkation Port**

```python
pd.crosstab(titanic['Embarked'],titanic['Survived']).plot(kind='bar')
plt.title('Survival by Embarkation Port')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.show()
```

**Average Fare by Passenger Class**

```python
avg_fare = titanic.groupby('Pclass')['Fare'].mean()

plt.bar(avg_fare.index, avg_fare.values)
plt.title('Average Fare by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Average Fare')
plt.show()
```

**Correlation Matrix**

```python
numeric_df = titanic.select_dtypes(include=['number'])

corr = numeric_df.corr()

plt.figure(figsize=(8,6))
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title('Correlation Matrix')
plt.show()
```

**Survival Rate by Age Group**

```python
survival_age = pd.crosstab(titanic['Age_Group'],titanic['Survived'])

survival_age.plot(kind='bar')
plt.title('Survival by Age Group')
plt.show()
```

1.Female passengers had a significantly higher survival rate than males.

2.First-class passengers were more likely to survive than third-class passengers.

3.Most passengers were young adults aged between 18 and 35.

4.Higher ticket fares were generally associated with better survival chances.

5.Family size showed some influence on passenger survival.

6.Passenger class had the strongest relationship with survival.
